```python
# now the problem is that the output of movies.iloc[0].genres is actually a string ,so we need to convert
# this string into list so that we can pass it into convert function to get the correct ans

#import ast
#ast.literal_eval
```

```python
import ast

def convert(obj):  # this obj is a string
    L = []
    for i in ast.literal_eval(obj):  # and in this step it will become a list
        L.append(i['name'])
    return L 
```

```python
def convert3(obj):  # this obj is a string
    L = []
    counter = 0
    for i in ast.literal_eval(obj):  # and in this step it will become a list
        if(counter != 3):
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# movies['crew'][0]
# this has many dictinaries but we want only that dictionary which has the job: director and in that dict
# we need the 'name' value

def fetch_director(obj):  # this obj is a string
    L = []
    for i in ast.literal_eval(obj):  # and in this step it will become a list
        if(i['job'] == 'Director'):
            L.append(i['name'])
            break   # as we want only one director
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# now our task is to check the similarity of movies....we will use vectorisation for that
# converting text to vector ...known as text vectorisation ....we will be using 'bag of words' technique
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english') # the 2nd attribute will
                                                                 # will remove the stop words of english language
```

```python
# CountVectorizer returns a scipi sparse matrix by default ,we will convert it to a numpy array
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()   # these are the top 5000 words
```

```python
# we will remove similar words by one word...like actor,actors...activity ,acitivities...etc
# we will use a process called steming ...example how it works---['loved','loving','love'] ==> ['love','love','love']
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():  # string converted to list 
        y.append(ps.stem(i))   #converting every word to stem and storing it in y
    
    string = " ".join(y)  # again converting this list into a string   
    return string    
```

```python
#example:
print(ps.stem('loved'))
print(ps.stem('love'))
print(ps.stem('dancing'))
print(ps.stem('dance'))
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# we have 4806 movies, that means we have 4806 vectors and every vector has 5000 numbers
# we have to calculate the distance of every movie with every other movie
# we wont be calculating the eucladian distance(tip to tip distance).....we will calculate the cosine distance (angle)
# if(angle == 0)  almost same,if it is 5,10...similar....90 - > dissimilar ...180 --> completely opposite
# hence for more similarity ,the distance has to be less
'''therefore  : similarity is inverse of distance'''
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# now the problem that comes with sorting is that first of all it gets sorted in ascending order,we want it to be 
# in descending order and 2ndly when we sort the data the index gets lost ....meaning like initially the first index of 
# similarity is similarity of the given movie with the first movie,the value at 2nd posn shows the sim of the given movie
# with the 2nd movie and when we sort this array ..these indexes will get lost ,we dont want that to happen
# we want to hold the value of those indexes
# so we will use enumeration
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]  # to take the first five values
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)     # now this will print the movies titles ..instead of the movie indexes of
                                           # top five similar movies
```

```python
recommend('Avatar')
```
